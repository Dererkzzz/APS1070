**Question 1**

```python
import matplotlib.pyplot as plt

def plot_product_family_sales(dataframe, product_family_list):


    fig, axs = plt.subplots(len(product_family_list), 1,figsize=(15, 5*len(product_family_list)))

    # Ensure axs is an array even if there's only one subplot
    if len(product_family_list) == 1:
        axs = [axs]

    for ax, family in zip(axs, product_family_list):
        # Extract the row corresponding to the current product family
        family_data = dataframe.loc[family]

        # Plot the data
        ax.plot(family_data.index, family_data.values, label=family)
        ax.set_title(family)
        ax.set_xlabel('Date')
        ax.set_ylabel('Sales')
        #ax.legend(loc="upper left")

    plt.tight_layout()
    plt.show()


```
